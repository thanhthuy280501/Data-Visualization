```python
import csv

# Path to your file - update this if your shortcut is named differently
file_path = '/content/drive/MyDrive/DATA/lego_setsHB.csv'

price_dict = {}
difficulty_dict = {}

try:
    with open(file_path, mode='r', encoding='utf-8') as file:
        # Using a counter to stop at 1000 lines
        for i, line in enumerate(file):
            if i >= 1000:
                break

            # Split the line by comma
            columns = line.strip().split(',')

            # Basic validation: ensure the line has enough columns
            # Assuming: ID is col 0, Price is col 1, Difficulty is col 2
            # (Adjust indices based on your specific CSV structure)
            if len(columns) >= 3:
                prod_id = columns[0]
                price = columns[1]
                difficulty = columns[2]

                # Populating the dictionaries
                price_dict[prod_id] = price
                difficulty_dict[prod_id] = difficulty

    print(f"Successfully processed {len(price_dict)} unique products.")

except FileNotFoundError:
    print("Error: The file was not found. Please ensure the 'DATA' folder shortcut is set up correctly in Colab.")
except Exception as e:
    print(f"An error occurred: {e}")

# Example check:
# print(list(price_dict.items())[:5])
```

```python
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/DATA/lego_setsHB.csv'
df_lego = pd.read_csv(file_path)

# Display summary information
print("--- DataFrame Info ---")
df_lego.info()

# Display descriptive statistics
print("\n--- Descriptive Statistics ---")
display(df_lego.describe())
```

```python
# Filter the DataFrame using boolean indexing for star_rating >= 4
star_rating_filter = df_lego['star_rating'] >= 4
high_rated_lego = df_lego[star_rating_filter]

# Display the filtered results
print(f"Number of Lego sets with star_rating >= 4: {len(high_rated_lego)}")
display(high_rated_lego.head())
```

```python
# find the number of Lego sets whose star_rating greater than or equal to 4. (Hint: consider using .count() on the specified column)
high_rated_count = high_rated_lego['star_rating'].count()
print(high_rated_count)
```

```python
# find the average list price for the Lego sets whose star_rating greater than or equal to 4. (Hint: consider specifying the list_price column after obtaining data and before taking the average)
average_price = high_rated_lego['list_price'].mean()
print(average_price)
```

```python
import pandas as pd
file_path = '/content/drive/MyDrive/DATA/olympic_medals.csv'
#Use the full file path variable to load the data
medals = pd.read_csv(file_path)
display(medals.head())
print('----------------------------------------------------------')

#Find the number of rows and columns, use the .shape
print('The number of rows, columns', medals.shape)
# Returns (rows, columns)
print('---------------------------------------------------------')


#Use the .dtypes to see the types for all columns
print(medals.dtypes)
print('----------------------------------------------------------')


#Use .unique() combined with len()
# Updated 'City' to 'Location'
unique_cities_count = len(medals['Location'].unique())
print('The uniques cities are', unique_cities_count)
print('---------------------------------------------------')


#Filter the dataframe for 'USA' and get the count
# Updated 'NOC' to 'Nationality'
usa_medals = medals[medals['Nationality'] == 'USA'].shape[0]
print('Medals the USA team won', usa_medals)
print('-----------------------------------------------------')


#Use the groupby() function as suggested to agregate the counts
# Updated 'NOC' to 'Nationality'
medals_per_nation = medals.groupby('Nationality')['Medal'].count()
print(medals_per_nation)
```
