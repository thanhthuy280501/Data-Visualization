```r
########################################
# Homework 2 - Introduction to Data Visualization
# Full R script for all questions
########################################

########## Question 1 ##########

# Create df1
df1 <- data.frame(
  Name = c('James', 'Paul', 'Richards', 'Marico', 'Samantha', 'Ravi',
           'Raghu', 'Richards', 'George', 'Ema', 'Samantha', 'Catherine'),
  State = c('Alaska', 'California', 'Texas', 'North Carolina', 'California',
            'Texas', 'Alaska', 'Texas', 'North Carolina', 'Alaska',
            'California', 'Texas'),
  Sales = c(14, 24, 31, 12, 13, 7, 9, 31, 18, 16, 18, 14)
)

# aggregate total Sales by State
q1_aggregate <- aggregate(df1$Sales, by = list(df1$State), FUN = sum)
q1_aggregate

# dplyr: group_by + summarise
library(dplyr)

q1_dplyr <- df1 %>%
  group_by(State) %>%
  summarise(sum_sales = sum(Sales))
q1_dplyr


########## Question 2 - WorldCupMatches.csv ##########

# Read the WorldCupMatches data
# Adjusted path to match your Google Drive location
worldcup <- read.csv("/content/drive/MyDrive/Colab Notebooks/WorldCupMatches.csv")

## (a) Size of the data frame
q2a_dim <- dim(worldcup)
q2a_dim

## (b) Statistical summary of the data
q2b_summary <- summary(worldcup)
q2b_summary

## (c) Number of unique locations
q2c_unique_locations <- length(unique(worldcup$City))
q2c_unique_locations

## (d) Average attendance
q2d_avg_attendance <- mean(worldcup$Attendance, na.rm = TRUE)
q2d_avg_attendance

## (e) Total goals by Home Team
q2e_goals_home <- worldcup %>%
  group_by(Home.Team.Name) %>%
  summarise(total_home_goals = sum(Home.Team.Goals, na.rm = TRUE))
q2e_goals_home

## (f) Average attendance by year
q2f_attendance_by_year <- worldcup %>%
  group_by(Year) %>%
  summarise(avg_attendance = mean(Attendance, na.rm = TRUE))
q2f_attendance_by_year


########## Question 3 - metabolites.csv ##########

# Read the metabolites data
# Adjusted path to match your Google Drive location
metab <- read.csv("/content/drive/MyDrive/Colab Notebooks/metabolite.csv")

## (a) Number of Alzheimers patients
q3a_alzheimers_count <- sum(metab$Diagnosis == "Alzheimers", na.rm = TRUE)
q3a_alzheimers_count

## (b) Number of missing values
q3b_missing_per_col <- colSums(is.na(metab))
q3b_missing_per_col

## (c) Remove rows with missing Dopamine
metab_no_na_dopa <- metab[!is.na(metab$Dopamine), ]

## (d) Replace missing values in c4-OH-Pro with median
# Using backticks for column name with hyphens
median_c4 <- median(metab_no_na_dopa$`c4-OH-Pro`, na.rm = TRUE)
metab_no_na_dopa$`c4-OH-Pro`[is.na(metab_no_na_dopa$`c4-OH-Pro`)] <- median_c4

## (e) Drop columns with > 25% missing
missing_pct <- colMeans(is.na(metab_no_na_dopa))
metab_filtered <- metab_no_na_dopa[, missing_pct <= 0.25]

dim(metab_filtered)
colnames(metab_filtered)
```
